### 1.3 RAG (Retrieval Augmented Generation) (10 poin)
a) Jelaskan konsep RAG dan mengapa penting dalam pengembangan AI aplikasi!

b) Sebutkan komponen utama dalam sistem RAG!

c) Apa perbedaan antara semantic search dan keyword search dalam konteks RAG?

**Jawaban:**
```
a) RAG (Retrieval-Augmented Generation) adalah sebuah teknik yang digunakan untuk meningkatkan kemampuan model bahasa besar (LLM) dalam menghasilkan respons yang akurat, informatif dan relevan, terutama jika menghadapi pertanyaan yang memerlukan pengetahuan spesifik atau informasi yang tidak ada dalam data pelatihan awal LLM. Secara sederhana, RAG bekerja dengan menambahkan fase "pencarian" atau "pengambilan informasi (retrieval)" sebelum LLM menghasilkan respons (generation). RAG sangat penting karena mampu mengatasi keterbatasan mendasar LLM (hanya mengandalkan pengetahuan sendiri dari data pelatihan). Penggunaan RAG ini ditujukan untuk mengurangi halusinasi dan meningkatkan akurasi faktual, artinya RAG menyediakan grounding faktual, memaksa LLM untuk menjawab berdasarkan bukti yang diambil dari sumber eksternal sehingga meningkatkan akurasi dan keandalan respons tanpa mengarang fakta atau informasi yang tidak benar.
b) Komponen utama dalam sistem RAG:
- Knowledge Base merupakan kumpulan data yang akan menjadi sumber informasi bagi LLM, berupa dokumen teks, halaman web, database, PDF, buku atau yang lainnya.
- Chunking (pembagian atau fragmentasi) merupakan proses memecah dokumen besar dari knowledge base menjadi potongan-potongan teks yang lebih kecil dan mudah di kelola.
- Embedding Model (model embedding atau vectorizer) merupakan model yang bertugas mengubah setiap chunk teks dari knowledge base (dan juga query pengguna) menjadi representasi numerik multidimensional yang disebut vektor embedding. Vektor-vektor ini menangkap makna semantik dari teks.
- Vector Database (Basis Data Vektor) / Vector Store / Vector Index merupakan basis data khusus yang dirancang untuk menyimpan dan mengidenks vektor embedding secara efisien. Komponen ini memungkinkan pencarian kesamaan vektor yang cepat antara vektor query pengguna dan vektor chunk dalam knowledge base.
- Retriever (Pengambil Informasi) merupakan komponen yang bertanggung jawab untuk mengambil query pengguna dan mengubahnya menjadi vektor menggunakan embedding model. Kemudian menggunakan vektor database untuk menemukan top-k (beberapa) chunk paling relevan dari knowledge base.
- Large Language Model (LLM) merupakan model utama yang akan menghasilkan respons akhir. LLM menerima prompt yang sudah diperkaya dengan informasi yang diambil oleh retriever.
- Prompt Engineering / Orchestration merupakan proses merangkai query asli pengguna dan chunk yang diambil menjadi prompt yang efektif untuk LLM. Hal ini sering melibatkan penambahan instruksi ke LLM tentang bagaimana menggunakan informasi yang diberikan dan format respons yang diinginkan.
c) Semantic Search (Pencarian Semantik) yaitu mencari dokumen atau teks berdasarkan makna di balik query pengguna, bukan hanya kecocokan kata kunci. hal  tersebut dilakukan dengan mengubah query dan semua dokumen menjadi vektor embedding dalam ruang berdimensi tinggi (menggunakan embedding model). Kemudian, sistem mencari vektor dokumen yang paling "mirip" atau "dekat" secara geometris dengan vektor query. Sedangkan Keyword Search (Pencarian Kata Kunci) yaitu mencari dokumen atau teks yang mengandung kata kunci atau frasa yang sama persis seperti yang ada dalam query pengguna.
```

### 2.3 Embedding & Vector Storage (15 poin)
Implementasikan proses embedding dan penyimpanan ke vector database:

#### TUGAS:
##### 1. Load sentence transformer model lokal (misal: 'all-MiniLM-L6-v2') atau dari Hugging Face dan boleh menggunakan model lain yang sesuai
##### 2. Generate embeddings untuk setiap text data pegawai
##### 3. Store embeddings ke ChromaDB
##### 4. Similarity search

````python
#Generate embeddings untuk setiap text data pegawai
#Store embeddings ke ChromaDB
# Tulis code Anda di sini:

# Load model lokal atau dari Hugging Face
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')  # ringan dan cepat
````

````python
import sqlite3
import pandas as pd

# Koneksi ke database SQLite
conn = sqlite3.connect("pegawai.db")

# Ambil data pegawai
df = pd.read_sql_query("SELECT * FROM pegawai", conn)

# Gabungkan informasi penting jadi satu string per pegawai
def to_text(row):
    return (
        f"{row['nama']} adalah seorang {row['jabatan']} yang bekerja di {row['unit_kerja']}. "
        f"Berpendidikan {row['pendidikan']} dengan spesialisasi {row['spesialisasi']}. "
        f"Memiliki pengalaman kerja selama {row['pengalaman_kerja']} tahun."
    )

df["text"] = df.apply(to_text, axis=1)
texts = df["text"].tolist()
ids = df["id"].astype(str).tolist()  # ID harus berupa string untuk ChromaDB
````

````python
# Buat embeddings dari teks
embeddings = model.encode(texts)

import chromadb
from chromadb.config import Settings

# Setup ChromaDB (lokal, in-memory untuk demo)
chroma_client = chromadb.Client(Settings(anonymized_telemetry=False))

# Buat collection
collection = chroma_client.get_or_create_collection(name="pegawai_embeddings")

# Masukkan dokumen & embedding ke ChromaDB
collection.add(
    documents=texts,
    embeddings=embeddings,
    ids=ids
)
````

````python
query = "orang yang berpengalaman di bidang teknologi"
query_embedding = model.encode([query])[0]

# Search di ChromaDB (top 3 hasil)
results = collection.query(
    query_embeddings=[query_embedding],
    n_results=3
)

# Tampilkan hasil
print("\n=== Hasil Similarity Search ===")
for doc, dist in zip(results['documents'][0], results['distances'][0]):
    print(f"Skor Kemiripan: {1 - dist:.4f}\n{doc}\n")
````

#### TUGAS: Test similarity search
##### Coba query seperti: "pegawai dengan keahlian IT", "staff dengan pendidikan S2"

````python
# Coba query seperti: "pegawai dengan keahlian IT", "staff dengan pendidikan S2"
# Tulis code Anda di sini:
# Pastikan model dan collection sudah siap (lanjutan dari sebelumnya)

test_queries = [
    "pegawai dengan keahlian IT",
    "staff dengan pendidikan S2"
]

for query in test_queries:
    print(f"\n🔍 Query: \"{query}\"")

    # Buat embedding dari query
    query_embedding = model.encode([query])[0]

    # Lakukan pencarian similarity
    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=3
    )

    # Tampilkan hasil
    for i, (doc, dist) in enumerate(zip(results['documents'][0], results['distances'][0]), start=1):
        print(f"\nHasil {i}: (Skor Kemiripan: {1 - dist:.4f})")
        print(doc)
````

````python
def generate_llm_response(prompt, context=None):
    try:
        query = f"""Gunakan informasi berikut untuk menjawab dalam bahasa Indonesia:\n\nKONTEKS:\n{context}\n\nPERTANYAAN:\n{prompt}""" if context else f"Jawablah dalam bahasa Indonesia:\n\n{prompt}"
        return qa_chain.invoke({"query": query})
    except Exception as e:
        return f"Terjadi kesalahan: {e}"
````
